```python
#FIRST PROGRAM FOR PRINT HELLO WORLD
print ("Hello World")

#FIRST PROGRAM OF VARIABLE AND TYPE
name = "Muhammad Humdan Ali"
print (type(name))

#FIRST METHOD OF CONCATENATION
name1 = "Muhammad"
name2 = "Humdan"
name3 = "Ali"
print (name1+" "+name2+" "+name3)

#SECOND METHOD OF CONCATENATION
name4 = "Muhammad"
name5 = "Humdan"
name6 = "Ali"
print (name4,name5,name6)
```

```python
#FIRST PROGRAM OF PRINTING A NUMBER
print (2)

#FIRST PROGRAM OF TYPE
print (type(2))

#FIRST PROGRAM OF VARIABLE FOR INTEGER
number1 = 100
print (number1)
print (type(number1))

#ARITHEMATIC PROGRAM WITHOUT VARIABLE
print(2+2)
print(4-2)
print(2*2)
print(4/2)
print(10//2)

#ARITHEMATIC OPERATIONS PROGRAM WITH VARIABLE
number1 = int(input("Enter the first number"))
number2 = int(input("Enter the second number"))
print (number1 + number2)
print (number1 - number2)
print (number1 * number2)
print (number1 / number2)
print (number1 // number2)
print (number1 ** number2)
```
